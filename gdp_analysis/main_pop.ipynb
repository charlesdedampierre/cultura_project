```python
import pandas as pd


data = pd.read_csv('results/population_interpolated.csv', index_col = [0])

# load regions and equivalent country_name
df_regions_culture = pd.read_csv('../environnement_data/ENS - Cultural Index - Countries Databases - consolidate_table.csv')
df_regions_culture = df_regions_culture[['region', 'iso_a3']].drop_duplicates()

df_iso = pd.read_csv('results/iso.csv', index_col = [0])


df_final = pd.merge(data,df_iso, on = 'country_name')
df_final = pd.merge(df_final, df_regions_culture, on = 'iso_a3')
df_final = df_final.groupby(['region', 'year'])['pop'].sum().reset_index()
df_final = df_final.rename(columns={'region':'region_name'})
df_final.to_csv('results/population_interpolated_regions.csv')
```
