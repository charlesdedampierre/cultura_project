```python
import git
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# Replace 'path/to/your/repo' with the path to your Git repository
repo_path = '../'

# Open the Git repository
repo = git.Repo(repo_path)
```

```python


# Create a directed graph using networkx
G = nx.DiGraph()

# Add nodes and edges based on commit history
for commit in repo.iter_commits():
    G.add_node(commit.hexsha, label=str(commit)[:8])
    if commit.parents:
        G.add_edge(commit.hexsha, commit.parents[0].hexsha)

# Visualize the commit tree
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=700, node_color='skyblue', font_size=8, font_color='black', font_weight='bold', arrowsize=15)

plt.show()
```
