```python
import sys

sys.path.append('../')

from functions_env import DB_PATH
import sqlite3
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
from matplotlib_venn import venn3


conn = sqlite3.connect(DB_PATH)

df_deathcity= pd.read_sql_query("SELECT * FROM individual_deathcity", conn)
df_info= pd.read_sql_query("SELECT * FROM individuals_main_information", conn)
df_info=df_info[['individual_wikidata_id','individual_name', 'birthyear', 'country_name']].drop_duplicates().dropna()
df_info['productivity_date'] =df_info['birthyear'] + 35
df_info = df_info[df_info['productivity_date'] < 2020]
print(df_info['country_name'])
```
