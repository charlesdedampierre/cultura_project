```python
import pandas as pd


df_ind = pd.read_csv('results/df_individuals_score.csv', index_col=[0])

# Manually clean the top 30
df = df_ind.sort_values(['region_name', 'score'], ascending=(False, False))
df = df.groupby(['individual_wikidata_id', 'individual_name', 'productive_year'])['region_name'].apply(lambda x: ' | '.join(x)).reset_index()
df
```

```python
df_only_indians = df_ind[df_ind['region_name']=='Indian world']
df_only_indians = df_only_indians.sort_values(['region_name', 'score'], ascending=(False, False)) 
df_only_indians = df_only_indians[30:]
df_only_indians = df_only_indians.groupby(['individual_wikidata_id', 'individual_name', 'score', 'productive_year'])['region_name'].apply(lambda x: ' | '.join(x)).reset_index()
df_only_indians = df_only_indians[['individual_wikidata_id', 'productive_year', 'individual_name', 'score', 'region_name']]
df_only_indians.to_clipboard(index=False)
```
